#### Hypothesis Testing:

#### Two-sample t-test: 

- H0(Null Hypothesis): There is no significant difference between the means monthly income  of males and females.
- H1(Alternate Hypothesis): There is signigficant difference between the male and female monthly income.

```python
monthly_income_male=df[df['Gender']=='Male']['MonthlyIncome'].values
monthly_income_female=df[df['Gender']=='Female']['MonthlyIncome'].values
```

```python
np.var(monthly_income_male)/np.var(monthly_income_female)
```

- The ratio between the two sample variance is less than 4. Hence we can assume equal variance among the samples.

```python
import scipy.stats as stats
 
# Perform the two sample t-test with equal variances
t_value,p_value=stats.ttest_ind(a=monthly_income_male, b=monthly_income_female, equal_var=True)
```

```python
if p_value > 0.05:
    print('The p-value {} is greater the significance level of 0.05.\n'.format(p_value))
    print('We fail to reject the null hypotheis -Implying there is no significant difference between the mean monthly incomes of male and female')
else:
    print('We reject the null hypothesis, indicating a significant difference between the mean monthly incomes of male and female')
```

### Model Building:

#### Splitting the data to features and target variables.

```python
X=df.drop('Attrition',axis=1)
y=df['Attrition']
```

```python
#Encode the target variable 
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
y=enc.fit_transform(y)
print(y)
```

```python
X.info()
```

```python
#The features df has got many catgeroical features and need to be encoded.
X=pd.get_dummies(X,columns=['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime'],drop_first=True)
```

```python
#Splitting the data into test and train data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

### Model Building using Random Forest Classifier:
#### Why choose RFC?
- The data set has multicollinearity present which has no impact on RFC. RFc doesn't make any assumptions about the underlying data.
- The data has many features, using RFC important features that impact the target can be chosen.

```python
#Using Random Forest Classifier to build the model
from sklearn.ensemble import RandomForestClassifier

 #The target is imbalanced,'class_weight'=balanced balances the weights for the target feature
rfc = RandomForestClassifier(class_weight='balanced', random_state = 4)   
rfc.fit(X_train, y_train)
```

#### Model Evaluation:

```python
y_pred = rfc.predict(X_test)
print('Random Forest Model Accuracy : ', rfc.score(X_test, y_test)*100, '%')
```

```python
from sklearn.metrics import classification_report

print("The classification report: \n\n {}".format(classification_report(y_test, y_pred)))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

acc_score = accuracy_score(y_test, y_pred)
prec_score = precision_score(y_test, y_pred)
rec_score = recall_score(y_test, y_pred)
f_score = f1_score(y_test, y_pred)
```

```python
labels = ['accuracy', 'precision', 'recall', 'f1']

evaluation = pd.DataFrame({labels[0] : acc_score, labels[1] : prec_score, labels[2] : rec_score, labels[3] : f_score}, index = ['Standard RFC'])
evaluation
```

#### Precision Recall Tradeoff:

- This probelm is based on attrtion value which focuses more on label '1' which implies employees leaving rather than employees staying and it is important to find the presence of attrition.
- By using precision recall trade-off, the recall can be improved focusing n attrition.

```python
def apply_threshold(probabilities, threshold):
    # +1 if >= threshold and -1 otherwise.
    elements=[1 if p > threshold else 0 for p in probabilities]

    return np.array(elements)
```

```python
probabilities = rfc.predict_proba(X_test)[:,1]
```

```python
predictions_with_30_treshold=apply_threshold(probabilities, 0.3)
print(len(predictions_with_30_treshold))
print(len(predictions_with_30_treshold[predictions_with_30_treshold==1]))
```

```python
threshold_values = np.linspace(0.2, 0.5, num=100)    #Assigning trehold values
probabilities = rfc.predict_proba(X_test)[:,1]
```

```python
precision_all = []
recall_all = []

for thresh in threshold_values:
    y_thresh=[1 if p > thresh else 0 for p in probabilities]
    precision_thresh=precision_score(y_test,y_thresh)
    recall_thresh=recall_score(y_test,y_thresh)
    
    precision_all.append(precision_thresh)
    recall_all.append(recall_thresh)
```

```python
dataset = pd.DataFrame({'threshold_values': threshold_values, 'precision_all': precision_all,'recall_all': recall_all}, columns=['threshold_values', 'precision_all','recall_all'])
```

```python
predictions_with_30_threshold = apply_threshold(probabilities, 0.25)
skplt.metrics.plot_confusion_matrix(y_test,predictions_with_30_threshold)
```

- The threshold of 0.25 enables to predict the class label '1' with better accuracy than the default 0.5 threshold.
- Out of 58, 17 were found to be misclassified which earlier was 37 out of 58. But the downside is that precision was traded off,earlier only 6 samples were misclassified out of 236.

#### Conclusion:
- The random forest classifier using SMOTE provided better evaluation metrics than the standard and hyper parameter tuned models.
- Still the recall score was a little off since the attrition presence couldnt been accuratly predicted.
- With the help of precision-recall tradeoff we were able to improve the recall score by trading off with precision. 
- The precision-recall must been traded off according to the problem at hand - If the company wants to focus more on retention of employess and finding the factors responsible, then precision must be given importance which the original model with SMOTE provided.
- If the company to promote or take steps for employee then to be certain they wont leave we have to use recall to improve recall score at the cost of precision.
